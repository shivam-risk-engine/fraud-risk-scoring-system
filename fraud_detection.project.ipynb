```python
claim_size = float(input("Enter claim amount: "))
frequency = int(input("Enter number of claims: "))
age = int(input("Enter age: "))

total = 0

if claim_size > 50000:
    total += 0.4

if frequency > 3:
    total += 0.3

if age < 30:
    total += 0.2


def rank(total):
    if total <= 0.2:
        return "Low Risk - Approve"

    elif total <= 0.5:
        return "Medium Risk - Review"

    elif total <= 0.8:
        return "High Risk - Investigate"

    else:
        return "Very High Risk - Strict Investigation"


result = rank(total)

print("Risk Score:", total)
print("Decision:", result)
```

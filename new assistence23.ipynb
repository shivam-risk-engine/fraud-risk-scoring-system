```python
import numpy as np

np.random.seed(43)

#  Number of customers
n_customers = 1000

#  Claim occurrence (0 = no claim, 1 = claim)
claim_occur = np.random.binomial(1, 0.3, n_customers)

#  Claim sizes (₹1000 to ₹10000)
size = np.random.randint(1000, 10000, n_customers)

#  Total loss per customer
losses = size * claim_occur


#  BASIC METRICS
total_loss = np.sum(losses)
mean = np.mean(losses)              # Expected Loss
std = np.std(losses)                # Risk (SD)
variance = np.var(losses)           # Variance
claim_percentage = np.mean(claim_occur) * 100

max_loss = np.max(losses)
min_loss = np.min(losses)

#  HIGH LOSS CASES
high_loss = losses[losses > 8000]


#  95% RANGE (Normal approximation)
upper_95 = mean + 2 * std
lower_95 = mean - 2 * std


#  VaR & CVaR (Tail Risk )
var_95 = np.percentile(losses, 95)

cvar_95 = np.mean(losses[losses > var_95])


#  PREMIUM CALCULATIONS
premium_model_1 = mean + std + 200            # Aggressive
premium_model_2 = mean + 0.3 * std            # Balanced
premium_model_3 = mean * 1.2                  # Market style
premium_model_4 = var_95 + 200                # Tail-based


#  OUTPUT
print("-------- RESULTS --------")
print("TOTAL LOSS:", total_loss)
print("EXPECTED LOSS (EV):", mean)
print("STANDARD DEVIATION (RISK):", std)
print("VARIANCE:", variance)
print("CLAIM FREQUENCY (%):", claim_percentage)

print("MAX LOSS:", max_loss)
print("MIN LOSS:", min_loss)
print("HIGH LOSS CASES (>8000):", len(high_loss))

print("\n95% RANGE (Approx):", lower_95, "to", upper_95)

print("\n------ TAIL RISK ------")
print("VaR (95%):", var_95)
print("CVaR (95%):", cvar_95)

print("\n------ PREMIUM MODELS ------")
print("Model 1 (EV + SD + 200):", premium_model_1)
print("Model 2 (EV + 0.3*SD):", premium_model_2)
print("Model 3 (EV * 1.2):", premium_model_3)
print("Model 4 (VaR + 200):", premium_model_4)
```
